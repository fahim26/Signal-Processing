```python
def Graph_Plot(x,y,c,xl,yl,flag):
  plt.figure(figsize = (15, 6))
  if flag == 1:
    plt.plot(x, y, c)
  else:
    plt.stem(x,y,c)
  plt.ylabel(yl)
  plt.xlabel(xl)
  plt.show()


sr = 110
ts = 1.0/sr
t = np.arange(0,1,ts)

freq = 2
x = 5*np.sin(2*np.pi*freq*t)
arr = [4,7,9]
ampl = [10,7,14]
i=0
for frq in arr:    
  x +=  ampl[i]*np.sin(2*np.pi*frq*t)
  i=i+1

Graph_Plot(t,x,'g',"Continuous Signal","Amplitude",1)
Graph_Plot(t,x,'b',"Discrete Signal","Amplitude",2)


```

```python
def DFT(x):
  N = len(x)
  n = np.arange(N)
  k = n.reshape((N, 1))
  e = np.exp(-2j * np.pi * k * n / N)
  
  X = np.dot(e, x)
  
  return X

X = DFT(x)

N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T 

def Plotting_Frequency_Domain(X):
  plt.figure(figsize = (14, 6))
  plt.stem(freq, abs(X), 'r', \
          markerfmt=" ", basefmt="-b")
  plt.xlabel('Freq (Hz)')
  plt.ylabel('DFT Amplitude |X(freq)|')
  plt.show()
Plotting_Frequency_Domain(X)
```

```python
def One_Sided_Values(N,freq):
  n_oneside = N//2
  f_oneside = freq[:n_oneside]
  X_oneside =X[:n_oneside]/n_oneside
  plt.figure(figsize = (16, 6))
  plt.subplot(121)
  plt.stem(f_oneside, abs(X_oneside), 'r', \
          markerfmt=" ", basefmt="-b")
  plt.xlabel('Freq (Hz)')
  plt.ylabel('DFT Amplitude |X(freq)|')

  plt.subplot(122)
  plt.stem(f_oneside, abs(X_oneside), 'r', \
          markerfmt=" ", basefmt="-b")
  plt.xlabel('Freq (Hz)')
  plt.xticks(range(14))
  plt.xlim(0, 14)
  plt.tight_layout()
  plt.show()
One_Sided_Values(N,freq)
```

```python
def IDFT(x):
  N =  np.size(x)
  X = np.zeros((N,),dtype=np.complex128)
  for m in range(0,N):
      for n in range(0,N): 
          X[m] += x[n]*np.exp(np.pi*2j*m*n/N)
  return X/N

x2 = IDFT(X)

Graph_Plot(t,x2,'b',"","Amplitude",2)
Graph_Plot(t,x2,'g',"","Amplitude",1)
```
