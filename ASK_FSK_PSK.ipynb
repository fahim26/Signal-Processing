```python
import numpy as np
import matplotlib.pyplot as plt
import itertools 
from scipy import signal
from matplotlib.pyplot import figure
```

```python
def Digital_Signal_Plot(digital_data,highest_v,lowest_v):
  a = len(digital_data)
  t = np.arange(0,(len(digital_data)+1)*0.1-0.1 , 0.0001)
  y = []
  flag = 0
  for i in digital_data :
    if i == 1 :
      for j in range(0,1000):    
        y.append(highest_v)
    else:
        for j in range(0,1000):
          y.append(lowest_v)
  return y
```

```python
def draw_vertical_line(digital_signal):
  a = len(digital_signal)
  tt = np.arange(0,(len(digital_signal)+1)*0.1,0.1)
  for i in tt:
    plt.axvline(x=i,linestyle="--",color='black')


def Plot_Graph(x,y,digital_data,colr,title,x_label,y_label,flag,super_title,):
  figure(figsize=(15, 5), dpi=80)
  plt.plot(x,y,colr ,linewidth = '3')
  if flag == 0:
    plt.suptitle(super_title,fontsize=20, color= 'Green', fontweight='bold')
  plt.title(title,fontsize=15, color= 'blue', fontweight='bold')
  plt.xlabel(x_label)
  plt.ylabel(y_label)
  plt.xticks(np.arange(0,(len(digital_data)+1)*0.1,0.1))
  plt.axhline(y=0, color='black', linestyle='-')
  draw_vertical_line(digital_data)
```

```python
def Amplitude_Shift_Keying(digital_data):
  carrier_frequency = 5*8
  digital_frequency = 5
  amplitude = 5
  t = np.arange(0,(len(digital_data)+1)*0.1-0.1 , 0.0001)
  carrier_signal = amplitude * np.sin(2*np.pi*carrier_frequency*t)
  digital_signal = Digital_Signal_Plot(digital_data , 10,0)
  modulated_signal = []
  modulated_signal.clear()
  modulated_signal = carrier_signal*digital_signal
  
  Plot_Graph(t,digital_signal,digital_data,'b','Digital Signal','Time','Amplitude',0,'ASK')
  Plot_Graph(t,carrier_signal,digital_data,'r','Carrier Signal' , 'Time' , 'Amplitude',1,'ASK')
  Plot_Graph(t,modulated_signal,digital_data,'g','Modulated Signal','Time' , 'Amplitude',1,'ASK')
```

```python
def Frequency_Shift_Keying(digital_data):
  carrier_frequency_1 = 5*2
  carrier_frequency_2 = 5*8
  digital_frequency = 5
  amplitude = 5
  t = np.arange(0,(len(digital_data)+1)*0.1-0.1 , 0.0001)
  carrier_signal_1 = amplitude * np.sin(2*np.pi*carrier_frequency_1*t)
  carrier_signal_2 = amplitude * np.sin(2*np.pi*carrier_frequency_2*t)
  digital_signal = Digital_Signal_Plot(digital_data , 10,0)
  modulated_signal = []
  modulated_signal.clear()
  i=0
  for value in t :
    if digital_signal[i] <= 0:
      modulated_signal.append(amplitude * np.sin(2*np.pi*carrier_frequency_1*value))
    else :
      modulated_signal.append(amplitude * np.sin(2*np.pi*carrier_frequency_2*value))
    i=i+1
    
  Plot_Graph(t,digital_signal,digital_data,'b','Digital Signal','Time','Amplitude',0,'FSK')
  Plot_Graph(t,carrier_signal_1,digital_data,'r','Carrier Signal 1' , 'Time' , 'Amplitude',1,'FSK')
  Plot_Graph(t,carrier_signal_2,digital_data,'r','Carrier Signal 2','Time' , 'Amplitude',1,'FSK')
  Plot_Graph(t,modulated_signal,digital_data,'g','Modulated Signal','Time' , 'Amplitude',1,'FSK')
```

```python
def Phase_Shift_Keying_2(digital_data):
  carrier_frequency_1 = 5*2
  carrier_frequency_2 = 5*4
  digital_frequency = 5
  amplitude = 5
  t = np.arange(0,(len(digital_data)+1)*0.1-0.1 , 0.0001)
  carrier_signal_1 = amplitude * np.sin(2*np.pi*carrier_frequency_1*t)
  carrier_signal_2 = amplitude * np.sin(2*np.pi*carrier_frequency_2*t)
  digital_signal = Digital_Signal_Plot(digital_data , 10,0)
  modulated_signal = []
  modulated_signal.clear()
  
  i=0
  for value in t :
    if digital_signal[i] <= 0:
      modulated_signal.append(-amplitude * np.sin(2*np.pi*carrier_frequency_1*value))
    else :
      modulated_signal.append(amplitude * np.sin(2*np.pi*carrier_frequency_2*value))
    i=i+1

  Plot_Graph(t,digital_signal,digital_data,'b','Digital Signal','Time','Amplitude',0,'PSK')
  Plot_Graph(t,carrier_signal_1,digital_data,'r','Carrier Signal 1' , 'Time' , 'Amplitude',1,'PSK')
  Plot_Graph(t,carrier_signal_2,digital_data,'r','Carrier Signal 2','Time' , 'Amplitude',1,'PSK')
  Plot_Graph(t,modulated_signal,digital_data,'g','Modulated Signal','Time' , 'Amplitude',1,'PSK')
```

```python
digital_signal = [1,0,0,1,1,1,0,0,1,0]
Amplitude_Shift_Keying(digital_signal)
Frequency_Shift_Keying(digital_signal)
Phase_Shift_Keying_2(digital_signal)

```
