```python
import numpy as np
import matplotlib.pyplot as plt
import itertools 
from matplotlib.pyplot import figure
```

```python
def plotting(total_figure,r,c,x,y,title,xlab,ylab,colour,flag):
  plt.subplot(total_figure, r,c)
  if flag == 0:
    plt.title(title,fontsize=15, color= 'blue', fontweight='bold')
  plt.plot(x,y,color=colour)
  plt.ylabel(ylab)
  plt.xlabel(xlab)
```

```python
def Frequency_Modulation(): 
  Carrier_amplitude = float(input('Carrier Amplitude: '))
  Carrier_frequency = float(input('Carrier Frequency: '))
  Message_amplitude = float(input('Message Amplitude: '))
  Message_frequency = float(input('Message Frequency: '))

  frequency_sensitivity = 1

  t = np.linspace(0,1,1000)
  # print(t)
  carrier_signal = Carrier_amplitude*np.cos(2*np.pi*Carrier_frequency*t)
  message_signal = Message_amplitude*np.cos(2*np.pi*Message_frequency*t)
  summ=np.zeros_like(message_signal)
  # print(carrier_signal)
  i=0
  # print("#####################################################################")

  for value in t:
    # print(value)
    summ[i] = np.sin(2*np.pi*Message_frequency*value)
    i=i+1

  B = (Message_amplitude*frequency_sensitivity)/Message_frequency
  # B=10
  Result = Carrier_amplitude*np.cos(2*np.pi*Carrier_frequency*t + B*summ )


  figure(figsize=(15, 15), dpi=80)
  plotting(3,1,1,t,message_signal,'Frequency Modulation','Modulator signal','Amplitude','blue',0)
  plotting(3,1,2,t,carrier_signal,'Frequency Modulation','Carrier signal','Amplitude','red',1)
  plotting(3,1,3,t,Result,'Frequency Modulation','FM signal','Amplitude','g',1)

Frequency_Modulation()
```
